```python
mysize=(2000/ np.array(range(1,101)))
mysize #**GENERATE 100 samples** of different sizes N \(N\>10 & N\<2000\)
```

# 1. ChiSquare Distribution

```python
chidist = [] #make a list to stuff things into, doesn't matter their type
for i in range(0, len(mysize)):
    chidist.append(np.random.chisquare(100, mysize[i]))
```

```python
chimeans = []
for i in range(0, len(mysize)):
    chimeans.append(chidist[i].mean())
```

```python
pl.scatter(mysize, chimeans)
pl.title('Chi Squared', fontsize=18)
pl.xlabel('Sample Size', fontsize = 18)
pl.ylabel('Sample Mean', fontsize = 18)
```

## Figure 1:
When the sample size is small (on the left side of the plot), there is a wide range of means (a high standard deviation), but those means are distributed according to the normal distribution (there are fewer means further from the mean mean than there are closer to it).  When the sample size is large (N = 2000) the mean of the sample is close to the mean of the means, which in this case is 100.  

```python
pl.hist(chimeans, bins=30)
pl.title('Chi Squared', fontsize=18)
pl.xlabel('Sample Mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

## Figure 2
Means from each sample, plotted in a histogram.  The means are centralized around the "population" mean, which is 100.  

# 2. Normal Distribution

```python
normdist = [] #make a list to stuff things into, doesn't matter their type
for i in range(0, len(mysize)):
    normdist.append(np.random.normal(100, size=mysize[i]))
```

```python
normmeans = []
for i in range(0, len(mysize)):
    normmeans.append(normdist[i].mean())
```

```python
pl.scatter(mysize, normmeans)
pl.title('Normal Distribution', fontsize=18)
pl.xlabel('Sample Size', fontsize = 18)
pl.ylabel('Sample Mean', fontsize = 18)
```

## Figure 3
The larger the sample size, the closer it's mean will be to the population mean. This scatter plot shows the mean for each of 100 normally distributed samples of varying sample sizes.

```python
pl.hist(normmeans, bins=30)
pl.title('Normal', fontsize=18)
pl.xlabel('Sample Mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

## Figure 4
The means of each 100 normally distributed samples are plotted.  The mode of the means is 100, as shown by the histogram.

## 5. Student's Choice: Wald
